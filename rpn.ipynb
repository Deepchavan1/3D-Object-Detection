```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import *
from keras.models import *
from keras.activations import *
import numpy as np
```

```python
def rgn(input_shape, num_anchors_per_cell):
    
    in_img = Input(shape = input_shape)
    # Block 1
    
    conv1_b1 = Conv2D(128, (3,3), padding='same', strides=(2,2))(in_img)
    bn1_b1 = BatchNormalization()(conv1_b1)
    relu1_b1 = ReLU()(bn1_b1)
    
    conv2_b1 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu1_b1)
    bn2_b1 = BatchNormalization()(conv2_b1)
    relu2_b1 = ReLU()(bn2_b1)
    
    conv3_b1 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu2_b1)
    bn3_b1 = BatchNormalization()(conv3_b1)
    relu3_b1 = ReLU()(bn3_b1)
    
    conv4_b1 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu3_b1)
    bn4_b1 = BatchNormalization()(conv4_b1)
    relu4_b1 = ReLU()(bn4_b1)
    
    deconv1 = Conv2DTranspose(256, (3,3), padding='same', strides=(1,1))(relu4_b1)
    dbn1 = BatchNormalization()(deconv1)
    dact1 = ReLU()(dbn1)
    
    
    # Block 2
    
    conv1_b2 = Conv2D(128, (3,3), padding='same', strides=(2,2))(relu4_b1)
    bn1_b2 = BatchNormalization()(conv1_b2)
    relu1_b2 = ReLU()(bn1_b2)
    
    conv2_b2 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu1_b2)
    bn2_b2 = BatchNormalization()(conv2_b2)
    relu2_b2= ReLU()(bn2_b2)
    
    conv3_b2 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu2_b2)
    bn3_b2 = BatchNormalization()(conv3_b2)
    relu3_b2 = ReLU()(bn3_b2)
    
    conv4_b2 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu3_b2)
    bn4_b2 = BatchNormalization()(conv4_b2)
    relu4_b2 = ReLU()(bn4_b2)
    
    conv5_b2 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu4_b2)
    bn5_b2 = BatchNormalization()(conv5_b2)
    relu5_b2 = ReLU()(bn5_b2)
    
    conv6_b2 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu5_b2)
    bn6_b2 = BatchNormalization()(conv6_b2)
    relu6_b2 = ReLU()(bn6_b2)
    
    deconv2 = Conv2DTranspose(256, (2,2), padding='same', strides=(2,2))(relu6_b2)
    dbn2 = BatchNormalization()(deconv2)
    dact2 = ReLU()(dbn2)
    
    
    # Block 3
    
    conv1_b3 = Conv2D(128, (3,3), padding='same', strides=(2,2))(relu6_b2)
    bn1_b3 = BatchNormalization()(conv1_b3)
    relu1_b3 = ReLU()(bn1_b3)
    
    conv2_b3 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu1_b3)
    bn2_b3 = BatchNormalization()(conv2_b3)
    relu2_b3= ReLU()(bn2_b3)
    
    conv3_b3 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu2_b3)
    bn3_b3 = BatchNormalization()(conv3_b3)
    relu3_b3 = ReLU()(bn3_b3)
    
    conv4_b3 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu3_b3)
    bn4_b3 = BatchNormalization()(conv4_b3)
    relu4_b3 = ReLU()(bn4_b3)
    
    conv5_b3 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu4_b3)
    bn5_b3 = BatchNormalization()(conv5_b3)
    relu5_b3 = ReLU()(bn5_b3)
    
    conv6_b3 = Conv2D(128, (3,3), padding='same', strides=(1,1))(relu5_b3)
    bn6_b3 = BatchNormalization()(conv6_b3)
    relu6_b3 = ReLU()(bn6_b3)
    
    deconv3 = Conv2DTranspose(256, (4,4), padding='same', strides=(4,4))(relu6_b3)
    dbn3 = BatchNormalization()(deconv3)
    dact3 = ReLU()(dbn3)

    
    
    
    output = tf.concat([dact1, dact2, dact3], axis=1)
    
    x = Conv2D(num_anchors_per_cell, (1,1), padding='same', strides=(1,1))(output)
    x = BatchNormalization()(x)
    prob_map = ReLU()(x)
    
        
    x = Conv2D(7*num_anchors_per_cell, (1,1), padding='same', strides=(1,1))(output)
    x = BatchNormalization()(x)
    reg_map = ReLU()(x)   
    
    prob_map = tf.transpose(prob_map, (0,2,3,1))
    
    reg_map = tf.transpose(reg_map, (0,2,3,1))
    
    prob_map = sigmoid(prob_map)
    
#     x = keras.models.Model(in_img, outputs=[prob_map, reg_map])
    return prob_map, reg_map

# model = rgn((128,128,3))
# model.summary()
```
